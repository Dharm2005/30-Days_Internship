# Feature Engineering on Fare Column

```python
df['SibSp'].value_counts()
```

```python
df[df['SibSp'] == 8]
```

```python
df[df['Ticket'] == 'CA. 2343']
```

```python
df1= pd.read_csv('test.csv')
```

```python
df1[df1['Name'].str.contains('Sage', case=False)]
```

```python
concat_df = pd.concat([df,df1])
```

```python
concat_df[concat_df['Ticket'] =='CA. 2343']
```

```python
concat_df['Ticket'].value_counts()
```

```python
concat_df['Individual_fare']=concat_df['Fare']/(concat_df['SibSp']+concat_df['Parch']+1)
concat_df['Individual_fare']
```

```python
concat_df['Individual_fare'].plot(kind='box')
```

```python
concat_df[['Individual_fare','Fare']].describe()
```

```python
# Check who pays the highest individual fare
concat_df[concat_df['Individual_fare'] == concat_df['Individual_fare'].max()]
```

```python
concat_df['Family_size'] = concat_df['SibSp'] + concat_df['Parch'] + 1 
```

```python
concat_df
```

```python
# Family Type
# 1 -> alone
# 2-4 -> small
# >5 -> large

def transform_family_size(num):
    
    if num == 1:
        return 'alone'
    elif num>1 and num<5:
        return 'small'
    else:
        return 'large'
    
```

```python
concat_df['Family_type'] = concat_df['Family_size'].apply(transform_family_size)
concat_df.sample(5)
```

```python
import matplotlib.pyplot as plt

survived = concat_df[concat_df['Survived'] == 1]['Family_type'].value_counts()
not_survived = concat_df[concat_df['Survived'] == 0]['Family_type'].value_counts()

survived.plot(kind='bar', label='Survived', alpha=0.7)
not_survived.plot(kind='bar', label='Not Survived',alpha=0.7)

plt.legend()
plt.xlabel('Family Type')
plt.ylabel('Count')
plt.title('Survival by Family Type')
plt.show()
```

```python
pd.crosstab(concat_df['Survived'],concat_df['Family_type'],normalize='columns').plot(kind='bar')
```
