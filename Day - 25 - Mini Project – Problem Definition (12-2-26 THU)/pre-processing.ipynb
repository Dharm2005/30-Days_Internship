## Converting ARFF to CSV

```python
import pandas as pd
```

```python
data_started = False
data_lines = []

with open("./data/chronic_kidney_disease_full.arff", "r") as file:
    for line in file:
        if line.strip().lower() == "@data":
            data_started = True
            continue
        if data_started:
            data_lines.append(line.strip())

# Creating dataframe
df = pd.DataFrame([row.split(",") for row in data_lines])

df.head()
```

```python
# Droping one last extra column

df = df.iloc[:, :25]
```

```python
df.shape
```

```python
# Assigning Columns name manually

columns = [
'age','bp','sg','al','su','rbc','pc','pcc','ba',
'bgr','bu','sc','sod','pot','hemo','pcv',
'wc','rc','htn','dm','cad','appet','pe','ane','class'
]

df.columns = columns
```

```python
# formating data

df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Replacing ? to NA to identify missing values

df.replace('?', pd.NA, inplace=True)
```

```python
df['class'].value_counts()
```

```python
# dropping one inconsistent label ('no')

df = df[df['class'] != 'no']
```

```python
# Saving CSV File

df.to_csv('./data/ckd_raw.csv', index=False)
```
